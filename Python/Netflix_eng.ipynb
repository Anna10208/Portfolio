### 3. Working with missing values:

- Count the number of missing/null values in the dataset

- Handle missing values — replace or remove them

```python
country_null = netflix_df['country'].isnull()
netflix_df[country_null].tail(30)
```

```python
# Count missing values for each column
missing_in_col = netflix_df.isnull().sum()
print("Missing values:\n", missing_in_col)

# Total number of missing values
missing_total = missing_in_col.sum()
print(f"Total number of missing values: {missing_total}")
```

```python
# Fill missing values in the 'director' and 'cast' columns with 'Unknown'

netflix_df[['cast', 'director']] = netflix_df[['cast', 'director']].fillna('Unknown')

# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

```python
# Review the values in the genre column and find entries like 'British TV Shows' and 'Korean TV Shows'

netflix_df['listed_in'].unique()
```

```python
# Function to fill missing values in 'country' based on 'British TV Shows' and 'Korean TV Shows' 
def fill_country_1(row):
    if pd.isna(row['country']):                     # Check if the value in the 'country' column is empty
        if 'Korean TV Shows' in row['listed_in']:   # If 'Korean TV Shows' is in the genre, return the value "Korea"
            return 'Korea'
        elif 'British TV Shows' in row['listed_in']:
            return 'United Kingdom'
    return row['country']                           # In other cases, the value remains unchanged
# Apply the function to the entire DataFrame
netflix_df['country'] = netflix_df.apply(fill_country_1, axis=1)  # Apply the function row by row to the column


# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

```python
# Fill the 'country' column based on matches in the 'director' column
def fill_country_2(row, df):
    if pd.isna(row['country']):  # Check if the cell is empty
        # Filter rows with matching director
        filtr_director = df['director'] == row['director']
        matching_director = df.loc[filtr_director, 'country'].dropna().unique()
        
        # If unique countries are found, return the first one
        if len(matching_director) > 0:
            return matching_director[0]
        
        # If no matches are found, return 'Unknown'
        return 'Unknown'
    
    # If the cell is not empty, return its value
    return row['country']

# Apply the function to the DataFrame
netflix_df['country'] = netflix_df.apply(fill_country_2, axis=1, df=netflix_df)

# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

```python
# Fill missing values in the added date column. 
# Since there are no missing values in release_year, we can fill the added date with the release year.

def fill_date_added(row):
    if pd.isna(row['date_added']):  # Check if the cell is empty
        # Convert release_year to a string representing the date at the end of the year
        return f"Deсember 31, {row['release_year']}"
    
    # If the cell is not empty, return its value
    return row['date_added']

# Apply the function to the DataFrame
netflix_df['date_added'] = netflix_df.apply(fill_date_added, axis=1)

# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

```python
# Check rows with missing values in duration and see that the value was mistakenly placed in the rating column
netflix_df [netflix_df['duration' ].isnull()]
```

```python
# Fill missing values in the added date column
def fill_duration(row):
    if pd.isna(row['duration']):  # Check if the cell is empty
        return row['rating'] # Return the value from the column 
    
    # If the cell is not empty, return its value
    return row['duration']

# Apply the function to the DataFrame
netflix_df['duration'] = netflix_df.apply(fill_duration, axis=1)

# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

```python
# Check the values in the rating column and see that entries containing 'min' need to be removed

netflix_df['rating'].unique()
```

```python
# Replace values containing 'min' with NaN in the 'rating' column

row_min = netflix_df['rating'].str.contains('min', na=False)  # The str.contains method checks whether the value in the 'rating' column contains the substring 'min'.
                                                              # na=False excludes possible missing values (NaN) from the check
netflix_df.loc[row_min, 'rating'] = np.nan   # Assign NaN

# Check the values in 'rating'
netflix_df['rating'].unique()
```

```python
# Replace NaN values in the 'rating' column with 'Unknown'

# Check where there are NaN values in the 'rating' column
row_nan = netflix_df['rating'].isnull()  

# Replace all NaN values in the 'rating' column with 'Unknown'
netflix_df.loc[row_nan, 'rating'] = 'Unknown'

# Check the number of missing values
missing_in_col = netflix_df.isnull().sum()
missing_in_col
```

### 4. Display the count of series for each rating in the 'rating' column

```python
# Create a filter for rows where 'type' equals 'TV Show'
row_show = netflix_df['type'] == 'TV Show'
# Extract values from the 'rating' column for rows matching the row_show filter and count the number of unique values in 'rating'
rating_counts = netflix_df.loc[row_show, 'rating'].value_counts().sort_values(ascending=False) # Sort in descending order
print(rating_counts)
```

### 5. Data Analysis

- In which year was the highest number of movies released?
- Which country produced the highest number of movies?
- What are the most popular genres? TOP 10 most popular genres.

```python
# In which year was the highest number of movies released?
# Filter to include only movies
movies = netflix_df[netflix_df['type'] == 'Movie']

# Count the number of movies by year
release_year_counts = movies['release_year'].value_counts()

# Determine the maximum number of movies released in a year
max_movies_count = release_year_counts.max()

# The years in release_year_counts are the indices; 
# extract the indices (years) corresponding to the maximum value in value_counts() and add them to a list
max_movies_years = list(release_year_counts[release_year_counts == max_movies_count].index)
max_movies_years_str = ",".join(map(str, max_movies_years)) # Apply the str function to each value in the list max_movies_years 
                                                           #to convert elements to strings, then join them into a single string using .join()

print(f"The highest number of movies {max_movies_count} was released in the following years: {max_movies_years_str}.")
```

```python
# Which country produced the highest number of movies?

# Filter to include only movies
movies = netflix_df[netflix_df['type'] == 'Movie']

# Count the number of movies by country
country_counts = netflix_df['country'].value_counts()

# Determine the maximum number of movies
max_movies = country_counts.max()

#  The indices in the country_counts series are countries; add to a list the countries with the maximum number of movies.
max_movies_countries = list(country_counts[country_counts == max_movies].index)
max_movies_countries_str = ",".join(max_movies_countries) #  Join them into a single string using .join()


print(f"The highest number of movies {max_movies} was released in the following countries: {max_movies_countries_str}.")
```

```python
# What are the most popular genres? Display the TOP-10 most popular genres.
# We will find popular genres for both movies and TV shows

# Count the number of movies and TV shows by genres (listed_in), sort descending, assign integer ranks

listed_in_rank = netflix_df['listed_in'].value_counts().sort_values(ascending=False).rank(ascending=False).astype(int).head(10)

print(f"TOP 10 most popular genres: {listed_in_rank}")
```

```python
# TOP 10 most popular genres with the count of movies and TV shows
# Count the number of movies and TV shows by genres (listed_in), sort descending and select TOP 10
top_genres = netflix_df['listed_in'].value_counts().head(10)

print("TOP 10 most popular genres:")
print(top_genres)
```

### 6. TOP 15 oldest movies and TOP 15 newest movies

```python
# TOP 15 oldest movies
# Filter only movies
movies = netflix_df[netflix_df['type'] == 'Movie']

# Sort the filtered DataFrame by release year ascending and get the TOP 15 oldest
alt_mowies = movies.sort_values('release_year',ascending=True).head(15)

print("TOP 15 oldest movies:")
alt_mowies
```

```python
# TOP 15 newest movies
# Since there are many movies released in 2021, to refine the ranking we will consider the date added to the site

# Convert 'date_added' to datetime format with error handling
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')

# Split the 'date_added' column into 'Year Added', 'Month Added', 'Day Added'
netflix_df['Year Added'] = netflix_df['date_added'].dt.year
netflix_df['Month Added'] = netflix_df['date_added'].dt.month
netflix_df['Day Added'] = netflix_df['date_added'].dt.day

# Filter only movies
movies = netflix_df[netflix_df['type'] == 'Movie']

# Sort data by specified columns in descending order
sorted_movies = movies.sort_values(
    ['release_year', 'Year Added', 'Month Added', 'Day Added'],
    ascending=[False, False, False, False]
)

# Display the TOP 15 newest movies
top_15_newest = sorted_movies.head(15)

print("TOP 15 newest movies:")
print(top_15_newest[['title', 'release_year', 'date_added']])

```
